Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
df2 = df.iloc[:,0:2]
df2.head()
```

```python
# Corr is only applied to non-categorical values 
df2.corr()
```

```python
sns.heatmap(df2.corr())
```

UNIVARIATE/BIVARIATE ANALYSIS

```python
sns.jointplot(x = 'tip', y = 'total_bill',data=df2)
```

```python
sns.jointplot(x = 'tip', y = 'total_bill',data=df2,kind='hex')
```

```python
# kind='reg' gives the best fit regression line(in the cneter) and the bell-like curves which is probability distribution function(at the corners)
sns.jointplot(x = 'tip', y = 'total_bill',data=df2,kind='reg')
```

For multivariate we use pairplot

```python
sns.pairplot(df2)
```

```python
# Here we are using df dataframe
sns.pairplot(df,hue='sex')
```

DIST PLOT

```python
# it helps us to check the distribution of the column feature
sns.displot(df['tip'].head())
df.head()
```

Categorical Plots

```python
# Count plot
sns.countplot(x='day',data=df)
```

```python
sns.countplot(y = 'smoker',data=df)
```

```python
# BAr plot
# unlike count plot we give two aixses here

sns.barplot(x='sex',y='smoker',data=df)  
```

```python
sns.barplot(y='smoker',x='sex', data=df)  
```

```python
# boxplot 
# it gives percentile

sns.boxplot(x='sex',y='smoker',data=df,palette='rainbow')
```
